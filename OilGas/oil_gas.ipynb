# Read data

```python
import pandas as pd

file_names = ['AnvBas_gas','AnvBas_oil','AnvGJ_gas','AnvGJ_oil','AnvP_gas','AnvP_oil','TilBas_gas','TilBas_oil','TilGJ_gas','TilGJ_oil','TilP_gas','TilP_oil']

dfs = {}
for fn in file_names:
    dfs[fn] = pd.read_excel("..\\Data\\OilGas\\" + fn + ".xlsx", index_col=0)  
```

```python
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt

fig, axs = plt.subplots(ncols=2, nrows=2, figsize=(10,10))

dfs['TilGJ_oil'].columns = ['t', 'Tilgang (GJ)', 'Oil']
dfs['AnvGJ_oil'].columns = ['t', 'Anvendelse (GJ)', 'Oil']
sns.lineplot(x="t", y="Oil", hue="Tilgang (GJ)", data=dfs['TilGJ_oil'], ax=axs[0,0])
sns.lineplot(x="t", y="Oil", hue="Anvendelse (GJ)", data=dfs['AnvGJ_oil'], ax=axs[0,1])

dfs['TilGJ_gas'].columns = ['t', 'Tilgang (GJ)', 'Gas']
dfs['AnvGJ_gas'].columns = ['t', 'Anvendelse (GJ)', 'Gas']
sns.lineplot(x="t", y="Gas", hue="Tilgang (GJ)", data=dfs['TilGJ_gas'], ax=axs[1,0])
sns.lineplot(x="t", y="Gas", hue="Anvendelse (GJ)", data=dfs['AnvGJ_gas'], ax=axs[1,1])

plt.savefig("figur_GJ.jpeg")




```

```python

fig, axs = plt.subplots(ncols=2, nrows=2, figsize=(10,10))

dfs['TilP_oil'].columns = ['t', 'Tilgang (pris)', 'Oil']
dfs['AnvP_oil'].columns = ['t', 'Anvendelse (pris)', 'Oil']
sns.lineplot(x="t", y="Oil", hue="Tilgang (pris)", data=dfs['TilP_oil'], ax=axs[0,0])
sns.lineplot(x="t", y="Oil", hue="Anvendelse (pris)", data=dfs['AnvP_oil'], ax=axs[0,1])

dfs['TilP_gas'].columns = ['t', 'Tilgang (pris)', 'Gas']
dfs['AnvP_gas'].columns = ['t', 'Anvendelse (pris)', 'Gas']
sns.lineplot(x="t", y="Gas", hue="Tilgang (pris)", data=dfs['TilP_gas'], ax=axs[1,0])
sns.lineplot(x="t", y="Gas", hue="Anvendelse (pris)", data=dfs['AnvP_gas'], ax=axs[1,1])

plt.savefig("figur_P.jpeg")
```
